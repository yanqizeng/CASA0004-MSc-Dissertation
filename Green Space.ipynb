```python
###  2. Convert Postcode to Coordinates  ###
```

```python
# Sheffield

from geopy.geocoders import Nominatim

# Read the data
data = pd.read_csv('GS_1_Sheffield_three periods.csv')  

# Display the first few rows of the dataset
# print(data.head())
```

```python
def postcode_to_coordinates(postcode):
    geolocator = Nominatim(user_agent="postcode_converter")
    location = geolocator.geocode(postcode)
    if location:
        return location.latitude, location.longitude
    else:
        return None, None

# Apply the geocoding function to the 'postcode' column and create new 'latitude' and 'longitude' columns
data['latitude'], data['longitude'] = zip(*data['postcode'].apply(postcode_to_coordinates))

# Display the updated dataset with coordinates
print(data.head())
```

```python
# Save the updated DataFrame to a new CSV file
data.to_csv('GS_2_Sheffield_with_coordinates.csv', index=False)  # Replace with desired file name

```

```python
###  3. Classification of homes using GIS technology  ###

### GIS technology: Homes were categorized according to the impact distance of the green space

### After GIS processing, I got data on homes within walking distance of green spaces
```

```python
# Cardiff

# 'GS_2_Cardiff_with_coordinates.csv' is all homes in Cardiff
# 'GS_3_Cardiff_with_coordinates_within400.csv' is the homes within distance of green space in Cardiff

# Read in data
df1 = pd.read_csv('GS_2_Cardiff_with_coordinates.csv')
df2 = pd.read_csv('GS_3_Cardiff_with_coordinates_within400.csv')
```

```python
# Assuming the common column for comparison is 'id' or any other common identifier
# Common column for comparison
common_column = 'id'

# Merge the DataFrames based on the common column
merged_df = pd.merge(df1, df2, on=common_column, suffixes=('_df1', '_df2'), how='outer')

# Create the 'distance400' column based on the comparison of 'priceper'
# Y: homes within distance of green space
# N: homes without distance of green space
merged_df['distance400'] = ['Y' if row['price_df1'] == row['price_df2'] else 'N' for _, row in merged_df.iterrows()]


# Save the result to a new CSV file
merged_df.to_csv('GS_4_Cardiffd_distance400_YorN.csv', index=False)

```

```python
###  4. Panel Regression  ###

###  In this model, a house within walking distance of the green space was represented as an entity. 
```

```python
# Cardiff

import statsmodels.api as sm


# Assuming the data is stored in a tab-separated values (TSV) format
Cardiff_400m = pd.read_csv('GS_4_Cardiff_distance400_YorN.csv')



Cardiff_400m['distance400_Y'] = (Cardiff_400m['distance400'] == 'Y').astype(int)


# Specify time periods
periods = [
    ('2019-01-01', '2020-03-23'),
    ('2020-03-24', '2021-04-12'),
    ('2021-04-13', '2022-12-12')
]

# Loop through time periods and run fixed effects regression
for start_date, end_date in periods:
    subset = Cardiff_400m[(Cardiff_400m['dateoftransfer'] >= start_date) & (Cardiff_400m['dateoftransfer'] <= end_date)]
    y = subset['price']
    X = subset[['distance400_Y']]  # Include other relevant control variables
    
    # Create panel data structure
    panel_data = pd.DataFrame({'y': y, 'distance400_Y': X['distance400_Y']})
    
    # Run fixed effects regression
    model = sm.OLS(panel_data['y'], sm.add_constant(panel_data['distance400_Y'])).fit()
    
    # Print regression results
    print(f"Time Period: {start_date} to {end_date}")
    print(model.summary())
    print("=" * 50)
```
